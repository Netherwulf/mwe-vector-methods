```python
import os
from tqdm.notebook import tqdm

import pandas as pd

tqdm.pandas()
```

```python
base_emb_dir = os.path.join('..', 'storage')

old_emb_filepath = os.path.join(base_emb_dir, 
                            'kgr10', 
                            'embeddings', 
                            'transformer', 
                            'sentences_containing_mwe_from_kgr10_group_0_embeddings_1_layers_incomplete_mwe_in_sent.tsv')

df_old_emb = pd.read_csv(old_emb_filepath, sep='\t', on_bad_lines='warn')
df_old_emb = df_old_emb[df_old_emb['is_correct'].isin(['0', '1', 0, 1])]
```

```python
# read number of rows quickly
length = sum(1 for row in open(old_emb_filepath, 'r'))

# define a chunksize
chunksize = 5000

# initiate a blank dataframe
df_old_emb_temp = pd.DataFrame()
print('reading old embeddings')
# tqdm context
with tqdm(total=length, desc="chunks read: ") as bar:
    # enumerate chunks read without low_memory (it is massive for pandas to precisely assign dtypes)
    for i, chunk in enumerate(pd.read_csv(old_emb_filepath, chunksize=chunksize, low_memory=False, on_bad_lines='warn')):
        # append it to df
        df_old_emb_temp = df_old_emb_temp.append(other=chunk)
        
        # update tqdm progress bar
        bar.update(chunksize)
```
